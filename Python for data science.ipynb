```python
marketing.columns = marketing.columns.str.strip()
```

```python
marketing['Income'] = marketing['Income'].str.replace(',', '')
marketing['Income'] = marketing['Income'].str.strip()
marketing['Income'] = marketing['Income'].str.replace('$', '', regex=False)
```

```python
marketing['Income'] = pd.to_numeric(marketing['Income'], errors='coerce')
marketing['Dt_Customer'] = pd.to_datetime(marketing['Dt_Customer'], errors='coerce')
```

```python
marketing.info()
```

```python
marketing['Income'].describe()

```

```python
marketing['Income'].isna().sum()
```

```python
marketing['Education'].unique()
```

```python
marketing['Education'] = marketing['Education'].replace({
    '2n Cycle': 'Master'
})
```

```python
marketing['Marital_Status'].unique()
```

```python
marketing['Marital_Status'] = marketing['Marital_Status'].replace({
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Single'
})
```

```python
marketing['Income'] = marketing['Income'].fillna(
    marketing.groupby(['Education', 'Marital_Status'])['Income'].transform('median')
)
```

```python
from datetime import datetime

current_year = datetime.now().year
marketing['Age'] = current_year - marketing['Year_Birth']
```

```python
marketing['Total_Children'] = marketing['Kidhome'] + marketing['Teenhome']
```

```python
marketing['Total_Spending'] = (
    marketing['MntWines'] +
    marketing['MntFruits'] +
    marketing['MntMeatProducts'] +
    marketing['MntFishProducts'] +
    marketing['MntSweetProducts'] +
    marketing['MntGoldProds']
)
```

```python
marketing['Total_Purchases'] = (
    marketing['NumWebPurchases'] +
    marketing['NumCatalogPurchases'] +
    marketing['NumStorePurchases']
)
```

```python
import matplotlib.pyplot as plt

marketing['Income'].hist(bins=30)
plt.show()
```

```python
Q1 = marketing['Income'].quantile(0.25)
Q3 = marketing['Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

marketing = marketing[(marketing['Income'] >= lower) & 
                        (marketing['Income'] <= upper)]
```

```python
education_order = {
    'Basic': 0,
    'Graduation': 1,
    'Master': 2,
    'PhD': 3
}

marketing['Education_Encoded'] = marketing['Education'].map(education_order)
```

```python
marketing['Age_Group'] = pd.cut(
    marketing['Age'],
    bins=[18, 35, 50, 100],
    labels=['Young', 'Middle', 'Senior']
)
```

```python
from scipy.stats import f_oneway

young = marketing[marketing['Age_Group']=='Young']['NumStorePurchases']
middle = marketing[marketing['Age_Group']=='Middle']['NumStorePurchases']
senior = marketing[marketing['Age_Group']=='Senior']['NumStorePurchases']

f_oneway(young, middle, senior)
```

```python
#Hypothesis 4:"Does the US outperform rest of world in total purchases?"

marketing['US_Flag'] = (marketing['Country'] == 'US').astype(int)
marketing.groupby('Country')['Total_Purchases'].mean()
us = marketing[marketing['Country']=='US']['Total_Purchases']
rest = marketing[marketing['Country']!='US']['Total_Purchases']

ttest_ind(us, rest, equal_var=False)
#Conclusion:There is no statistically significant evidence that the United States outperforms other countries in total purchase volume
```
